List vs Numpy - Time Taken (Numpy helps in faster Operations)

```python
from time import process_time
```

Time taken by List

```python
from IPython.utils import process
python_list = [i for i in range(100000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

Time taken by Numpy array

```python
python_numpy = np.array([i for i in range(100000)])

start_time = process_time()

python_numpy += 5

end_time = process_time()

print(end_time - start_time)
```

Numpy Arrays

```python
#list
list1 = [1,2,3,4,5,6]
print(list1)
print(type(list1))

#numpy array

np_array = np.array(list1)
print(np_array)
print(type(np_array))
```

```python
a = np.array([1,2,3,4,5,6,7])
print(a.shape)
b = np.array([(1,2,3,4),(1,2,3,4)])
print(b)
print(b.shape)
```

```python
a = np.array([1,2,3,4], dtype= float)
print(a)
```

Initial Placeholders in numpy arrays

```python
#create a numpy arrays of zeros ans ones
x = np.zeros((4,4), dtype= int)
print(x)
print()
x = np.ones((4,4))
print(x)
```

```python
x = np.full((4,4),7)
print(x)
```

```python
#identity matrix

a = np.eye(4,dtype = int)
print(a)
```

```python
#create a numpy array with random values
b = np.random.random((3,4)) #between 0-1
print(b)
#random integers from 10 - 100
b = np.random.randint(10,100,(4,4))
print(b)
```

```python
#array of evenly spaced values --> specifying the number of values required
d = np.linspace(10,30,5)
print(d)
#array of evenly spaced values --> specifying the number of steps, [start, stop)
d = np.arange(10,30,5)
print(d)
```

```python
#convert a list to a numpy array
list1 = [1,2,3,4,5,6]
e = np.array(list1)
f = np.asarray(list1)
print(e,type(e))
print(f,type(f))
#The key difference is that e is guaranteed to be a new NumPy array, whereas f
# might reuse the existing list if it was already a NumPy array.
```

Analysing a numpy array

```python
c = np.random.randint(10,90,(5,5))
print(c.shape)
print(c.ndim) #dimensions
print(c.size) #size of the array
print(c.dtype) #data type of the array
```

Mathematical Operations on a numpy array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9,10]
print(list1 + list2) #concatenate the lists

a = np.random.randint(1,10,(3,3))
b = np.random.randint(11,20,(3,3))
print(a)
print(b)

print(a+b) #it adds the values
print(np.add(a,b)) # also adds
```

Array manupulation

```python
#two ways to transpose a matrix
g = np.random.randint(10,40,(3,5))
print(g.T)
g = np.transpose(g)
print(g)
```

```python
#reshaping a array
a = np.random.randint(1,9,(2,4))
print(a)
print(a.shape)
b = a.reshape(4,2)
print(b)
print(b.shape)
```
